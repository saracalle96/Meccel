```python
import pymcel as pc
import numpy as np
import matplotlib.pyplot as plt
```

```python
sistema = [
    dict(m = 1, r = [0, 0, 0], v = [0, 0, 0]),
    dict(m = 1, r = [1, 0, 0], v = [0, 1, 0]) #ponemos la velocidad en z 
]
```

```python
ts = np.linspace(0, 10, 100)
rs, vs, rps, vps, constantes = pc.ncuerpos_solucion(sistema, ts)
```

```python
rs.shape #hay dos particulas, 100 tiempos, 3 coordenadas x, y, z es un cubo de datos, cada capa es una particula, cada columna el tiempo y cada fila las coordenadas
```

```python
plt.plot(rs[0, :, 0], rs[0, :, 1], label='Cuerpo 1') #primera particula, cualquier tiempo, coordenadas x, y
plt.plot(rs[1, :, 0], rs[1, :, 1], label='Cuerpo 2') #segunda particula, cualquier tiempo, coordenadas x, y
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trayectorias de los cuerpos')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(rps[0, :, 0], rps[0, :, 1], label='Cuerpo 1') #primera particula, cualquier tiempo, coordenadas x, y
plt.plot(rps[1, :, 0], rps[1, :, 1], label='Cuerpo 2') #segunda particula, cualquier tiempo, coordenadas x, y
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trayectorias de los cuerpos')
plt.legend()
plt.grid()
plt.show()
```
