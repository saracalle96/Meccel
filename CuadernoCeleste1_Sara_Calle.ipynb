```python
import pymcel as pc
import numpy as np
```

```python
m_tierra = pc.constantes.M_earth
m_sol = pc.constantes.M_sun
```

Posiciones de la Tierra y el Sol:

```python
tabla, jd, X = pc.consulta_horizons(id ='399', location = '@SSB', epochs = '2026-2-19') #baricentro del sistema solas SSB, 339 codigo de la Tierra
r_tierra = X[0 : 3]
```

```python
tabla, jd, X = pc.consulta_horizons(id = '0', location = '@SSB', epochs = '2026-2-19')
r_sol = X[0 : 3]
```

Calculamos el vector entre ellos:

```python
r_tierra_sol = r_tierra - r_sol
```

La fuerza entre ellos:

```python
F_tierra_sol = pc.constantes.G * m_tierra * m_sol * r_tierra_sol / np.linalg.norm(r_tierra_sol)**3
F_tierra_sol
```

Calculemos la fuerza que ejerce el Sol sobre la Luna:

```python
m_luna = pc.constantes.mu_moon / pc.constantes.G

tabla, jd_moon, X_moon = pc.consulta_horizons(id='301', location='@SSB', epochs='2026-2-19')
r_luna = X_moon[0:3]

r_luna_sol = r_luna - r_sol

F_luna_sol = pc.constantes.G * m_luna * m_sol * r_luna_sol / np.linalg.norm(r_luna_sol)**3
F_luna_sol
```

Calculemos la fuerza que ejerce la Tierra sobre la Luna:

```python
r_luna_tierra = r_luna - r_tierra

F_luna_tierra = pc.constantes.G * m_luna * m_tierra * r_luna_tierra / np.linalg.norm(r_luna_tierra)**3
F_luna_tierra
```

Calculamos la magnitud de ambas fuerzas:

```python

magnitud_F_luna_sol = np.linalg.norm(F_luna_sol)
magnitud_F_luna_tierra = np.linalg.norm(F_luna_tierra)

print(f"Magnitud de la fuerza del Sol sobre la Luna: {magnitud_F_luna_sol:.3e} N")
print(f"Magnitud de la fuerza de la Tierra sobre la Luna: {magnitud_F_luna_tierra:.3e} N")

# Comparación
if magnitud_F_luna_sol > magnitud_F_luna_tierra:
    print(f"La fuerza del Sol sobre la Luna es {magnitud_F_luna_sol / magnitud_F_luna_tierra:.2f} veces mayor que la fuerza de la Tierra sobre la Luna.")
elif magnitud_F_luna_tierra > magnitud_F_luna_sol:
    print(f"La fuerza de la Tierra sobre la Luna es {magnitud_F_luna_tierra / magnitud_F_luna_sol:.2f} veces mayor que la fuerza del Sol sobre la Luna.")
else:
    print("Ambas fuerzas tienen magnitudes similares.")
```
