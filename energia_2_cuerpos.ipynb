```python
import pymcel as pc
import numpy as np
import matplotlib.pyplot as plt
```

## Sistema dos cuerpos

```python
sistema = [ 
    dict(m=1, r=np.array([0, 0, 0]), v=np.array([0, 0, 0])),
    #dict(m=0.01, r=np.array([1, 0.0]), v=np.array([0.1, 0]))
    dict(m=0.01, r=np.array([1, 0, 0]), v=np.array([0, np.sqrt(2*G*1/1.0), 0]))
]
```

## Calculamos la energia total E = K + U

```python
k = 0.5 * sistema[1]['m'] * np.linalg.norm(sistema[1]['v'])**2 + 0.5 * sistema[0]['m'] * np.linalg.norm(sistema[0]['v'])**2
```

```python
G = 1 # En unidades canonicas
U = - G * sistema[0]['m'] * sistema[1]['m'] \
    / np.linalg.norm(sistema[0]['r'] - sistema[1]['r'])
```

```python
E = k + U
E
```
